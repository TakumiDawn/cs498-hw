```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os, glob
filenames_list = []
for dirname, _, filenames in os.walk('./'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
    filenames_list = filenames
```

```python
# df1 = pd.read_csv(filenames[1])
# df1 = pd.read_csv("./2011.csv")

all_files = glob.glob("*.csv")
print(all_files)
df_list = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    print(df.shape)
    df_list.append(df)

frame = pd.concat(df_list, axis=0, ignore_index=True)
print()
print(frame.shape)
```

```python
#Q1
avg_likes = round(frame['nlikes'].mean())
avg_likes
```

```python
#Q2
avg_replies = round(frame['nreplies'].mean())
avg_retweets = round(frame['nretweets'].mean())

sum_avg = avg_likes+avg_replies+avg_retweets
avg_view = sum_avg/0.05
avg_view
```

```python
#Q3
# Based on https://postcron.com/en/blog/social-media-image-dimensions-sizes/
# 
# I assume average photo size to be 0.2 MB, and average video size to be 3MB.
avg_tweet_size = round(frame['tweet'].str.len().mean())
avg_tweet_size
# content avg_tweet_size = 92 bytes = 9.2e-5 MB
```

```python
photo_cell_len = frame['photos'].apply(len)
# number of tweets with photos
nphotos = photo_cell_len[photo_cell_len>2].count()
nphotos
```

```python
# number of tweets with videos
video_col = frame['video']
nvideos = video_col[video_col>0].count()
nvideos
```

```python
frame.shape[0]
```

```python
avg_amount_data = avg_tweet_size*10**-5 + (nphotos*0.2+nvideos*3)/frame.shape[0]
avg_amount_data
```

```python
#Q4
total_avg_data = avg_amount_data *avg_view
total_avg_data
```

```python
#Q5
# Based on https://www.statista.com/statistics/282087/number-of-monthly-active-twitter-users/#:~:text=As%20of%20the%20first%20quarter,daily%20active%20users%20(mDAU)
# I assume Twitter averaged 330 million monthly active users
total_views  = avg_view * frame.shape[0]
total_views
```

```python
#total minutes in 11 years
total_min = 11 * 365*24*60
total_min
```

```python
view_per_min = total_views/total_min
view_per_min
```

```python
#Q6
# assume 10 tweets are displayed to a user per view
total_data_per_min = view_per_min * avg_amount_data *10
total_data_per_min
```
